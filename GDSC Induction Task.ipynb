```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("training_data.csv")
df.head()
```

```python
inputs = df
```

```python
from sklearn.preprocessing import LabelEncoder
le_oly = LabelEncoder()
le_scho = LabelEncoder()
le_sch = LabelEncoder()

le_fav = LabelEncoder()
le_pro = LabelEncoder()
le_med = LabelEncoder()

le_car = LabelEncoder()
le_act = LabelEncoder()
le_fan = LabelEncoder()

le_won = LabelEncoder()
le_ho = LabelEncoder()
```

```python
inputs['Olympiad_Participation_n'] = le_oly.fit_transform(inputs['Olympiad_Participation'])
inputs['Scholarship_n'] = le_scho.fit_transform(inputs['Scholarship'])
inputs['School_n'] = le_sch.fit_transform(inputs['School'])

inputs['Fav_sub_n'] = le_fav.fit_transform(inputs['Fav_sub'])
inputs['Projects_n'] = le_pro.fit_transform(inputs['Projects'])
inputs['Medals_n'] = le_med.fit_transform(inputs['Medals'])

inputs['Career_sprt_n'] = le_car.fit_transform(inputs['Career_sprt'])
inputs['Act_sprt_n'] = le_act.fit_transform(inputs['Act_sprt'])
inputs['Fant_arts_n'] = le_fan.fit_transform(inputs['Fant_arts'])

inputs['Won_arts_n'] = le_won.fit_transform(inputs['Won_arts'])
inputs['Predicted Hobby'] = le_ho.fit_transform(inputs['Predicted Hobby'])
```

```python
inputs.head()
inputs.columns
```

```python
inputs_n = inputs.drop(['Won_arts','Fant_arts','Act_sprt','Career_sprt','Medals','Projects','Fav_sub','School','Scholarship', 'Olympiad_Participation'],axis='columns')
```

```python
inputs_n.head()
```

```python
inputs = inputs_n.drop('Predicted Hobby',axis='columns')
```

```python
target = inputs_n['Predicted Hobby']
```

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression(solver='sag',multi_class='ovr', random_state = 0)  #Logical regression
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
svm = SVC(gamma='auto', kernel='rbf', random_state = 0)          #SVM
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators=200,random_state = 0)     #Random Forest
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
from sklearn import tree           #decision tree
model = tree.DecisionTreeClassifier(random_state = 0)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
